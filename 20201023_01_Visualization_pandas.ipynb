```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
url = r'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv'
```

```python
drinks = pd.read_csv(url)
drinks.head()
```

```python
# drinks.rename(columns={'country':'country','beer_serving':beer...})
drinks.columns = ['country', 'beer', 'split', 'wine', 'liters', 'continent'] # 전체를 한 번에 바꿀 때는 이렇게 해두 됨
drinks.head()
```

```python
drinks.info()
```

```python
drinks.set_index('country', inplace=True)
```

```python
drinks.head()
```

```python
# 1. Line Chart
```

```python
plt.figure(figsize=(5,3))
drinks['beer'].plot()
```

```python
drinks.plot()
```

```python
plt.figure(figsize=(10,5))
drinks.sort_values('beer')['beer'].plot()
```

```python
# 2. bar chart - y값의 크기
```

```python
plt.figure(figsize=(10,5))
drinks['beer'].plot(kind='bar')
```

```python
plt.figure(figsize=(10,5))
drinks['beer'].sort_values(ascending=False)[:50].plot(kind='bar')
```

```python
# 대륙별 국가평균 맥주소비량을 bar chart로 표시
```

```python
drinks.groupby('continent')['beer'].mean().plot(kind='bar')
```

```python
drinks.groupby('continent').mean().plot(kind='bar')
```

```python
drinks.groupby('continent').mean().plot(kind='bar', stacked='True')
```

```python
# histogram - 구간에 포함된 데이터의 수를 bar로 그림
```

```python
drinks['beer'].plot(kind='hist', bins=50)
plt.xlabel('Beer Servings')
plt.ylabel('Frequency')
plt.title('Histogram of Beer')
```

```python
# scatter - x좌표, y좌표
```

```python
drinks.plot(kind='scatter', x='beer', y='wine')
```

```python
pd.plotting.scatter_matrix(drinks)
plt.show()
```

```python
drinks.plot(kind='box')
```
