```python
# 1. BasePay 평균은 ?
sal['BasePay'].mean()
```

```python
# 2. OvertimePay의 최대값은?
sal['OvertimePay'].max()
```

```python
# 3. 'JOSEPH DRISCOLL'의 직업은?
sal[sal['EmployeeName']=='JOSEPH DRISCOLL']['JobTitle']
```

```python
# 4. 'TotalPayBenefits' 최고금액을 수령하는 사람의 이름은?
sal[sal['TotalPayBenefits']==sal['TotalPayBenefits'].max()]['EmployeeName']
```

```python
# 5. 연도별 BasePay의 평균금액은?
sal.groupby('Year')['BasePay'].mean()
```

```python
# 6. 'JobTitle'의 종류는?
len(sal['JobTitle'].unique())
```

```python
# 7. 'JobTitle'의 종류 중 가장 많은 JobTitle 상위 5개는 무엇인가?
sal.groupby('JobTitle').count().sort_values('Id', ascending=False).head().index
```

```python
job_count = {}
for j in sal.JobTitle:
    if j in job_count:
        job_count[j] +=1
    else:
        job_count[j] = 1
sorted(job_count.items(), key = lambda x: x[1], reverse=True)
```
