```python
pop_seoul = pd.read_excel('./01. population_in_Seoul.xls', usecols = [1,3,6,9,13],
                         skiprows=2)
pop_seoul.head(5)
```

```python
# 열이름 바꾸기
pop_seoul.rename(columns = {'자치구':'구별', '계':'인구수', '계.1':'한국인', '계.2':'외국인', '65세이상고령자':'고령자'}, inplace=True)
```

```python
pop_seoul.head() # rename할 때 inplace=True 옵션을 줘야 원본을 바꿈 안 그러면 그냥 그렇게 출력만 해주고 끝
```

```python
pop_seoul.info()
```

```python
pop_seoul.tail()
```

```python
pop_seoul.drop(26, axis=0, inplace=True)
```

```python
pop_seoul.tail()
```

```python
# 구별 외국인 비율 상위 5개 구
(pop_seoul['외국인']/pop_seoul['인구수'])*100
```

```python
pop_seoul['외국인비율'] = (pop_seoul['외국인']/pop_seoul['인구수'])*100
pop_seoul.sort_values('외국인비율', ascending=False).head()
```

```python
# 고령자 비율 상위 5개 구
pop_seoul['고령자비율'] = (pop_seoul['고령자']/pop_seoul['인구수'])*100
pop_seoul.sort_values('고령자비율', ascending=False)[:5]
```

```python
cctv_seoul.rename(columns = {'기관명':'구별'}, inplace = True)
```

```python
cctv_seoul.head()
```

```python
# CCTV가 가장 많은 구 상위 다섯
```

```python
cctv_seoul.sort_values('소계',ascending=False)[:5]
```

```python
# 2016년도 CCTV증가율이 가장 높은 구
cctv_seoul['2015년도 누적 cctv수'] = cctv_seoul[['2013년도 이전', '2014년', '2015년']].sum(axis=1)
```

```python
cctv_seoul
```

```python
cctv_seoul['2016년 증가율'] = cctv_seoul['2016년']/cctv_seoul['2015년도 누적 cctv수']
```

```python
cctv_seoul.sort_values('2016년 증가율', ascending=False)[:5]
```

```python
# 1인당 CCTV수(인당 cctv비율)가 가장 높은 상위 5개 구
```

```python
pop_cctv = pd.merge(cctv_seoul, pop_seoul, on='구별', how='inner')
pop_cctv.head()
```

```python
pop_cctv.drop(['2013년도 이전','2014년','2015년', '2016년', '2015년도 누적 cctv수', '2016년 증가율'], axis=1, inplace=True)
```

```python
pop_cctv.head()
```

```python
(pop_cctv['소계']/pop_cctv['인구수'])*100
```

```python
pop_cctv['인당 cctv수'] = pop_cctv['소계']/pop_cctv['인구수']
```

```python
pop_cctv.sort_values('인당 cctv수', ascending=False)[:5]
```

```python
# 1인당 cctv에 가장 큰 영향을 미치는 요인
pop_cctv.corr() # 상관계수 계산
```
