```python
# 1. 고객들의 평균구매금액은?
ecom['Purchase Price'].mean()
```

```python
# 2. 최고구매금액과 최소구매금액은?
print(ecom['Purchase Price'].max())
print(ecom['Purchase Price'].min())
```

```python
# 3. 영어사용자의 구매금액의 합계?
ecom['Language'].unique()
```

```python
ecom[ecom['Language']=='en']['Purchase Price'].sum()
```

```python
ecom.groupby('Language')['Purchase Price'].sum().loc['en']
```

```python
# 4. 사용언어별 평균구매금액은?
ecom.groupby('Language')['Purchase Price'].mean()
```

```python
# 5. 'Job'이 'Lawyer'인 사람의 숫자?
sum(ecom['Job']=='Lawyer') # sum 씌워버리면 True 개수만 세서 반환해줌
```

```python
count = 0
for j in ecom['Job']:
    if 'Lawyer' in j:
        count += 1
count
```

```python
count = 0
for j in ecom['Job']:
    if 'lawyer' in j.lower():
        count += 1
count
```

```python
# 6. 'AM'과 'PM'의 각각의 구매건수?
ecom.groupby('AM or PM')['Purchase Price'].count()
```

```python
# 7. Lot이 '90 WT'인 곳에서 구매한 구매금액은?
ecom[ecom['Lot']=='90 WT']['Purchase Price']
```

```python
# 8. credit card num이 '4926535242672853'인 사람의 email은?
ecom[ecom['Credit Card']==4926535242672853]['Email']
```

```python
# 9. 카드회사가 'American Express'이고 구매금액이 $95 이상인 구매고객 수?
ecom[ecom['CC Provider'] == 'American Express']
ecom[ecom['Purchase Price']>=95]
```

```python
ecom[(ecom['CC Provider'] == 'American Express') & (ecom['Purchase Price']>=95)]
```

```python
# 10. credit card 만료일이 2025년도인 카드를 소지한 구매자의 수?
ecom['CC Exp Date'][0].split('/')[1]
```

```python
def extract_mon(x):
    return x.split('/')[1]
```

```python
ecom['CC Exp Date'].apply(extract_mon)
```

```python
sum(ecom['CC Exp Date'].apply(lambda x:x.split('/')[1]) == '25')
```

```python
# 11. email 도메인 별 사용자 수는?
ecom['Email'][0].split('@')[1]
```

```python
ecom['Email'].apply(lambda x: x.split('@')[1]).value_counts()
```

```python
ecom['Email domain'] = ecom['Email'].apply(lambda x: x.split('@')[1])
ecom.groupby('Email domain')['Email'].count().sort_values(ascending=False)
```

```python
# 12. browser 종류별 이용자수?
ecom['Browser Info'][0].split('/')[0]
```

```python
ecom['Browser Info'].apply(lambda x: x.split('/')[0]).value_counts()
```
