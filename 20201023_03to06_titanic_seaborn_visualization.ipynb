```python
# 1. pcclass의 클래스별 평규연령 구하기
```

```python
df.groupby('pclass')['age'].mean()
```

```python
# 2. age의 null값을 pclass의 평균나이로 대체
```

```python
age_list = []
for i in zip(df['age'], df['pclass']):
    if np.isnan(i[0]):
        if i[1]==1:
            age_list.append(38)
        elif i[1]==2:
            age_list.append(29)
        else:
            age_list.append(25)
    else:
        age_list.append(i[0])
age_list
```

```python
df['age'] = age_list
```
