```python
plt.figure(figsize = (3,2)) # 도화지 만들기
x = [1,2,3,4,5]
y = [6,7,8,9,10]

plt.plot(x,y)
```

```python
plt.figure(figsize = (3,2))

x = [1,2,3,4,5]
y = [i**2 for i in x]

x1 = [1,2,3,4,5]
y1 = [6,7,8,9,10]

plt.plot(x,y)
plt.plot(x1,y1)
```

```python
x = [1,2,3,4,5]
y = [i**2 for i in x]

x1 = [1,2,3,4,5]
y1 = [6,7,8,9,10]

plt.plot(x,y)
plt.plot(x1,y1) #주피터 노트북은 도화지 만들기 안 해도 알아서 만들어줌
```

```python
plt.figure(figsize = (3,2))

x1 = range(50)
y1 = [np.cos(value) for value in x1]

y2 = [np.sin(value) for value in x1]

plt.plot(x1, y1)
plt.plot(x1, y2)
```

```python
plt.figure(figsize = (5,3))

plt.subplot(1,2,1)
plt.plot(x1, y1, color='red', linestyle = 'dotted')

plt.subplot(1,2,2)
plt.plot(x1, y2, color='green', linestyle ='dashed')
```

```python
x1 = range(100)
y1 = [value for value in x1]

x2 = range(100)
y2 = [value+50 for value in x2]

plt.plot(x1, y1, c='#87F5F5', ls = '-.', label = 'Line1')
plt.plot(x2, y2, c = '#6DD66D', ls = ':')

plt.text(50,60, 'Line 1')
plt.annotate('Line 2', xytext=(30,110), xy = (50,100), \
             arrowprops=dict(facecolor = 'yellow', shrink=0.1))

plt.legend(fancybox=True, shadow=True, loc = 'upper right') # 라벨 표시하게끔 해주는 거

plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Line Chart')

plt.savefig('LineChart.jpg')
```

### Bar Plot

```python
x = np.arange(0,8,2)
x
```

```python
x
```

```python
plt.bar(x, y0, color='blue', width = 0.5)
plt.bar(x, y1, color='green', width = 0.5, bottom=np.sum(data[:0], axis=0))
plt.bar(x, y2, color='r', width = 0.5, bottom=np.sum(data[:1], axis=0))
```
