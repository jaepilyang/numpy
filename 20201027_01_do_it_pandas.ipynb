### 챕터 6 - concat, merge

```python
df1 = pd.read_csv('../numpy_pandas/do ti pandas/data/concat_1.csv')
df2 = pd.read_csv('../numpy_pandas/do ti pandas/data/concat_2.csv')
df3 = pd.read_csv('../numpy_pandas/do ti pandas/data/concat_3.csv')
```

```python
# concat을 통해서 세 개의 dataframe을 세로방향으로 가로방향으로 병합
```

```python
pd.concat([df1, df2, df3], axis=0)
```

```python
pd.concat([df1, df2, df3], axis=1)
```

```python
# df1에 새로운 행 [a4, b4, c4, d4]를 추가
s = pd.Series(['a4','b4','c4','d4'])
s
```

```python
s = pd.DataFrame([s.tolist()], columns=['A','B','C','D'])
s
```

```python
pd.concat([df1, s])
```

```python
# ignore_index
```

```python
pd.concat([df1, df2, df3], axis=0, ignore_index=True)
```

```python
pd.concat([df1, df2, df3], axis=0).reset_index()
```

```python
pd.concat([df1, df2, df3], axis=0).reset_index(drop=True)
```

```python
pd.concat([df1, df2, df3], axis=1)
```

```python
pd.concat([df1, df2, df3], axis=1, ignore_index = True) # 칼럼명 정리해줌
```

```python
# concat시 컬럼명이나 인덱스 이름이 서로 다른 경우
```

```python
df2.columns = ['E', 'F', 'G', 'H']
df3.columns = ['A', 'C', 'F', 'H']
```

```python
df3
```

```python
pd.concat([df1, df2, df3])
```

```python
# 이름이 동일한 컬럼만 concat
```

```python
pd.concat([df2, df3], join = 'inner')
```

```python
pd.concat([df1, df3], join = 'inner')
```

```python
# 인덱스가 서로 다른 df를 row방향으로 concat
```

```python
df1.index = [0,1,2,3]
df2.index = [4,5,6,7]
df3.index = [0,2,5,7]
```

```python
# 인덱스를 기준으로 동일한 인덱스에 concat 된다.
# => 완전하게 concat을 하기 위해서는 index를 통일시켜야 한다.

pd.concat([df1, df2, df3], axis=1)
```

```python
pd.concat([df1, df3], axis=1, join='inner')
```

```python
pd.concat([df2, df3], axis=1, join='inner')
```
